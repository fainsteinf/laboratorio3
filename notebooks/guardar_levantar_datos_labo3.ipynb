## Guardar y levantar datos - Laboratorio 3

En esta notebook tenemos instrucciones estándar para el guardado de datos.

**Guardar una única adquisición**

Hago un ejemplo sintético, tienen que adaptarlo a los nombres de las variables que hayan elegido.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

#creo valores ficticios 
freq = 5
tiempo = np.linspace(0, 1, 1000)    
voltaje_1 = np.sin(tiempo*2*np.pi*freq)
voltaje_2 = (1/freq) * np.sin(tiempo*2*np.pi*freq)

############################## A partir de aca ###########################

#los acomodamos en un dataframe de pandas
df = pd.DataFrame({'Time':tiempo,'V1':voltaje_1,'V2':voltaje_2})

#si queremos guardar la frecuencia (se agrega una columna extra)
df["Frequency"] = freq

#Aca podemos decir que la frecuencia y el tiempo sean indices
df.set_index(["Frequency", "Time"])
```

```python
#Ahora lo guardamos
save_folder = 'path/'
df.to_csv(save_folder+"data_test.csv", index=False, header=True)
```

**Levantar los datos y graficar**

```python
df_ff = pd.read_csv(save_folder+"data_test.csv", index_col=["Frequency", "Time"])

df_ff
```

```python
#Graficar 
tiempo = df["Time"].values
v1 = df["V1"].values

plt.figure(figsize=(4,1))
plt.plot(tiempo, v1)
plt.show()
```

## Guardado de datos de un loop 

Si hacemos un barrido en frecuencia y queremos guardar los datos en un unico archivo.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

# Example data
frequencies = [100, 200, 300]
N = 1000  # Number of time points per frequency

# Create empty lists to store all data
all_frequencies = []
all_times = []
all_voltage1 = []
all_voltage2 = []

for f in frequencies:
    time = np.linspace(0, 1, N) 
    voltage1 = np.sin(2 * np.pi * f * time)  
    voltage2 = np.cos(2 * np.pi * f * time)  
    
    all_frequencies.extend([f] * len(time))
    all_times.extend(time)
    all_voltage1.extend(voltage1)
    all_voltage2.extend(voltage2)

# Create DataFrame in one step
df = pd.DataFrame({
    "Frequency": all_frequencies,
    "Time": all_times,
    "Voltage1": all_voltage1,
    "Voltage2": all_voltage2
})

df
```

```python
#Guardado
save_folder = 'path/'
df.to_csv(save_folder+"data_test.csv", index=False, header=True)
```

```python
#Levantar los datos
df_ff = pd.read_csv(save_folder+"data_test.csv", index_col=["Frequency", "Time"])

df_ff
```
